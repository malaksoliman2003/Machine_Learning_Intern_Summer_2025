```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/malak/OneDrive - Alexandria University/Documents/ML/first_intern_project.csv")
```

```python
import seaborn as sns
sns.countplot(x="market segment type", hue="booking status", data=dataset)
plt.title("Booking Status by Market Segment")
plt.xlabel("Market Segment")
plt.ylabel("Number of Bookings")
plt.legend(title="Booking Status")
plt.show()
```

```python
# Create Crosstab
ct = pd.crosstab(dataset['repeated'], dataset['booking status'], normalize='index')

ct.plot(kind='bar', stacked=True, color=["#364188", "#B40404"], figsize=(6,6))
plt.title("Booking Status vs Repeated")
plt.xlabel("Repeated or Not")
plt.ylabel("Number of Guests")
plt.show()
```

```python
# Create Crosstab
ct = pd.crosstab(dataset['special requests'], dataset['booking status'], normalize='index')

ct.plot(kind='bar', stacked=True, color=["#513688", "#B47104"], figsize=(6,6))
plt.title("Booking Status vs Special Requests")
plt.xlabel("Special Requests")
plt.ylabel("Number of Guests")
plt.show()
```

```python
dataset['P'] = dataset['P-C'] > dataset['P-not-C']

dataset['P'] = dataset['P'].replace({True: 'Higher', False: 'Lower'})

ct = pd.crosstab(dataset['P'], dataset['booking status'], normalize='index')

ct.plot(kind='bar', stacked=True, color=["#182C74", "#DF9F9F"], figsize=(6,6))
plt.title("Booking Status vs Previous Cancellations Percentage")
plt.xlabel("Percentage of Previous Cancellations")
plt.ylabel("Proportion of Guests")
plt.show()

```

```python
# and it has columns: 'lead_time' and 'booking_status'
plt.figure()
sns.kdeplot(
    data=dataset,
    x='average price ',
    hue='booking status',
    common_norm=False,     # don't normalize across groups
    fill=False             # remove fill to match your plot
)

plt.title('Average Price  by Cancellation Status')
plt.xlabel('average price')
plt.ylabel('Density')
plt.legend(title='booking status')
plt.show()


```

```python

dataset['Percentage of Adults'] = dataset['number of adults'] > dataset['number of children']
ct = pd.crosstab(dataset['Percentage of Adults'], dataset['booking status'], normalize='index')

ct.plot(kind='bar', stacked=True, color=["#BAC98D", "#B40404"], figsize=(6,6))
plt.title("Booking Status vs Dominance of Adults")
plt.xlabel("Dominance of Adults")
plt.ylabel("Number of Guests")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='booking status', y='P-not-C', data=dataset,estimator= np.mean, color='green')

plt.xlabel('Booking Status (0 = Not Cancelled, 1 = Cancelled)')
plt.ylabel('Average Number of Previous Cancellations')
plt.title('Previous Cancellations by Current Booking Status')
plt.show()



```

```python

# and it has columns: 'lead_time' and 'booking_status'
plt.figure()
sns.kdeplot(
    data=dataset,
    x='lead time',
    hue='booking status',
    common_norm=False,     # don't normalize across groups
    fill=False             # remove fill to match your plot
)

plt.title('Lead Time Density by Cancellation Status')
plt.xlabel('lead time')
plt.ylabel('Density')
plt.legend(title='booking status')
plt.show()
```

```python
dataset['type of meal'] = dataset['type of meal'].replace(
    ['Meal Plan 1', 'Meal Plan 2', 'Meal Plan 3'],
    'Meal Plan'
)
ct = pd.crosstab(dataset['type of meal'], dataset['booking status'], normalize='index')

ct.plot(kind='bar', stacked=True, color=["#BAC98D", "#B40404"], figsize=(6,6))
plt.title("Booking Status vs Meal Plan")
plt.xlabel("Meal Plan")
plt.ylabel("Number of Guests")
plt.show()
```
